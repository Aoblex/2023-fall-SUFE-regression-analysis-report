```r
library(survival)

# 创建一个示例数据集
# 假设有一个生存时间变量 time 和一个二元的事件指示变量 status
data <- data.frame(
  time = c(5, 10, 15, 20, 25),
  status = c(1, 1, 0, 1, 0),
  covariate1 = c(1, 2, 3, 4, 5),
  covariate2 = c(0, 1, 1, 0, 1)
)

# 拟合 Cox 比例风险回归模型
cox_model <- coxph(Surv(time, status) ~ covariate1 + covariate2, data = data)

# 打印 Cox 回归模型的摘要信息
summary(cox_model)
```
