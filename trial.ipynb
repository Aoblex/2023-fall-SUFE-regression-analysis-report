```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv(r'.\lung-cancer.csv',sep=",")
data["Survival months"] = data["Survival months"] + 1
data.head()
```

```python
#data = data[data["Year of death recode"] != "Alive at last contact"]
censored_data = data[data["Year of death recode"] == "Alive at last contact"]
uncensored_data = data[data["Year of death recode"] != "Alive at last contact"]
num_censored = len(censored_data)
num_uncensored = len(uncensored_data)
print(f"censored: {num_censored}")
print(f"uncensored: {num_uncensored}")
max_survival_months = max(uncensored_data["Survival months"])
plt.hist(x= uncensored_data.loc[:,"Survival months"],
         bins=[i for i in range(max_survival_months + 1)])
plt.show()
```
